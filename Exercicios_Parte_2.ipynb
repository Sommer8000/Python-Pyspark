##Exercicio 14

```python
import random
# amostra aleatoriamente 50 números do intervalo 0...500
random_list = random.sample(range(500), 50)

random_list.sort()
tam = len(random_list) #Tamanho da lista para referencia
print(random_list)
print('Valor Minímo:', random_list[0] , '\nValor Máximo:', random_list[tam-1])

soma=0

for x in range (0, tam):
  soma=soma+random_list[x]

media=soma/tam
print('Media:', media)

mediana=(random_list[24]+random_list[25])/2
print('Mediana:', mediana)
```
